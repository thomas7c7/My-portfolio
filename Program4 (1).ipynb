```python
#Grade ID: F2705 
#Program 4
#I shorted the data by 1000 for this assignment since there were a lot.

import pandas as pd
from textblob import TextBlob

#Loading the data 
df = pd.read_csv('DisneylandReviews.csv', encoding="latin-1")
df = df[:1000] #Here 

# the sentiment of the star rating
def classify_sentiment(x):
    if x <= 2:
        return 'negative'
    elif x == 3:
        return 'neutral'
    else:
        return 'positive'

df['Rating Sentiment'] = df['Rating'].apply(classify_sentiment)

# sentiment polarity using TextBlob
def sentiment_polarity(x):
    blob = TextBlob(x)
    return blob.sentiment.polarity

df['Text Sentiment'] = df['Review_Text'].apply(sentiment_polarity)

# part 4 --
def classify_sentiment_text(x):
    if x < -0.05:
        return 'negative'
    elif x > 0.05:
        return 'positive'
    else:
        return 'neutral'

df['Text Sentiment Classified'] = df['Text Sentiment'].apply(classify_sentiment_text)

# Determine star rating sentiment -- true / false
def star_text_sentiment_agreement(row):
    if row['Rating Sentiment'] == row['Text Sentiment Classified']:
        return True
    else:
        return False

df['Star-Text Sentiment Agreement'] = df.apply(star_text_sentiment_agreement, axis=1)

# table that summarizes the overall sentiment agreement of the reviews
total_num_reviews = df.shape[0]

num_agree = df[df['Star-Text Sentiment Agreement'] == True].shape[0]

num_disagree = df[df['Star-Text Sentiment Agreement'] == False].shape[0]

percent_agree = (num_agree/total_num_reviews) * 100

percent_disagree = (num_disagree/total_num_reviews) * 100

print('Overall Star-Sentiment Agreement')
print('Number of reviews:', total_num_reviews)
print('Number agree:', num_agree)
print('Number disagree:', num_disagree)
print('Percent agree: {:.2f}%'.format(percent_agree))
print('Percent disagree: {:.2f}%'.format(percent_disagree))

# table that summarizes the sentiment agreement for the reviews by star rating
df_grouped = df.groupby('Rating').agg({
    'Star-Text Sentiment Agreement': ['sum', 'count'],
}).reset_index()

df_grouped.columns = ['Rating', 'Num Agree', 'Total Reviews']

df_grouped['Num Disagree'] = df_grouped['Total Reviews'] - df_grouped['Num Agree']

df_grouped['% Agree'] = (df_grouped['Num Agree']/df_grouped['Total Reviews']) * 100

df_grouped['% Disagree'] = (df_grouped['Num Disagree']/df_grouped['Total Reviews']) * 100

print('\nReview Stars | Total Reviews | Num Agree | % Agree | Num Disagree | % Disagree')
for index, row in df_grouped.iterrows():
    print(row['Rating'], '\t', row['Total Reviews'], '\t', row['Num Agree'], '\t', 
          '{:.2f}%'.format(row['% Agree']), '\t', row['Num Disagree'], '\t', 
          '{:.2f}%'.format(row['% Disagree']))
```
